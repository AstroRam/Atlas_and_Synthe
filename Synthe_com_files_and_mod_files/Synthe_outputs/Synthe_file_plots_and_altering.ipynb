```python
import pandas as pd
data = pd.read_csv('5000T44377g_solar_scaled.txt', sep='\s+', header=None)
data = pd.DataFrame(data)
import matplotlib.pyplot as plt
x = data[0]
y = data[1]
fig, ax = plt.subplots()

ax.plot(x,y,'g')
ax.set_xlim(2500,10500)
ax.set_xlabel('Wavelength in Armstrong')
ax.set_ylabel('flux in erg/sec/cm2/Angstrong')

ax.grid(True)
plt.title('2500to10500Synthe100000resolution5000T44377')
plt.show()
```
